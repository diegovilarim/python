```python
df = pd.concat([df_logica, df_portugues, df_java, df_python, df_formacao], axis = 1)
"""O parâmetro axis especifica a direcção ao longo da qual um método ou função em particupar é aplicado num DataFrame.
O parâmetro eixo=0 representa que a função é aplicada em função de coluna, e axis=1 significa que a função é aplicada em função da linha."""
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score

treino_x, teste_x, treino_y, teste_y = train_test_split(x, y, test_size = 0.25)
print("Treinaremos com %d elementos e testaremos com %d elementos" % (len(treino_x), len(teste_x)))

modelo = LinearSVC()
modelo.fit(treino_x, treino_y)
previsoes = modelo.predict(teste_x)

acuracia = accuracy_score(teste_y, previsoes) * 100
print("A acurácia foi %.2f%%" % acuracia)
```

**Precisamos de um teste replicável, por isso, vamos conhecer o SEED. É uma definição importante para que nosso teste seja replicável. Faz com que a escolha dos dados não seja tão aleatória.**

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score

SEED = 20

treino_x, teste_x, treino_y, teste_y = train_test_split(x, y, random_state = SEED, test_size = 0.25)
print("Treinaremos com %d elementos e testaremos com %d elementos" % (len(treino_x), len(teste_x)))

modelo = LinearSVC()
modelo.fit(treino_x, treino_y)
previsoes = modelo.predict(teste_x)

acuracia = accuracy_score(teste_y, previsoes) * 100
print("A acurácia foi $.2f%%" % acuracia)
```
